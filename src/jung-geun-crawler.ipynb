```python
from bs4 import BeautifulSoup
import urllib.request

html = urllib.request.urlopen('https://bbs.ruliweb.com/best/humor/hit')

soup = BeautifulSoup(html, "html.parser")

link = soup.find_all("li", { "class" : "item"})

# print(soup.prettify())

for m in link:
    if 'title' in m.a.attrs:
        print(m.a['title'])
        print(m.a['href'])
```

```python
from bs4 import BeautifulSoup
import re
import requests
import urllib
import os


headers = {'Content-Type': 'application/json; charset=utf-8',
"User-Agent" : "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/74.0.3729.169 Chrome/74.0.3729.169 Safari/537.36"}

url = 'https://gall.dcinside.com/mgallery/board/lists?id=aoegame&exception_mode=recommend'

r = requests.get(url, headers = headers)

soup = BeautifulSoup(r.text, 'html.parser')

link = soup.find_all("td", { "class" : "gall_tit"})


for m in link:
    print(m.a.text)
    print(m.a['href'])

```
